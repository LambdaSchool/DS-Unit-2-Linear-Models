```python
# Load data downloaded from https://srcole.github.io/100burritos/
import pandas as pd
df = pd.read_csv(DATA_PATH+'burritos/burritos.csv')
```

```python
# Derive binary classification target:
# We define a 'Great' burrito as having an
# overall rating of 4 or higher, on a 5 point scale.
# Drop unrated burritos.
df = df.dropna(subset=['overall'])
df['Great'] = df['overall'] >= 4
```

```python
# Clean/combine the Burrito categories
df['Burrito'] = df['Burrito'].str.lower()

california = df['Burrito'].str.contains('california')
asada = df['Burrito'].str.contains('asada')
surf = df['Burrito'].str.contains('surf')
carnitas = df['Burrito'].str.contains('carnitas')

df.loc[california, 'Burrito'] = 'California'
df.loc[asada, 'Burrito'] = 'Asada'
df.loc[surf, 'Burrito'] = 'Surf & Turf'
df.loc[carnitas, 'Burrito'] = 'Carnitas'
df.loc[~california & ~asada & ~surf & ~carnitas, 'Burrito'] = 'Other'
```

```python
# Drop some high cardinality categoricals
df = df.drop(columns=['Notes', 'Location', 'Reviewer', 'Address', 'URL', 'Neighborhood'])
```

```python
# Drop some columns to prevent "leakage"
df = df.drop(columns=['Rec', 'overall'])
```

```python
df
```

```python
df['Great'] = df['Great'].astype(int)
df['Great'].head()
```

```python
# need to format date and time for test/train data

df['Date'] = pd.to_datetime(df['Date'], infer_datetime_format=True)
df['Date']
```

- [ ] Do train/validate/test split. Train on reviews from 2016 & earlier. Validate on 2017. Test on 2018 & later.


```python
# train/validate/test split
# train: review <= 2016
# validate: review == 2017
# test: review == 2018

train =  df[(df['Date'] <= '2016-12-31')]
val = df[(df['Date'] >= '2017-01-01') & (df['Date'] <= '2017-12-31')]
test = df[(df['Date'] >= '2018-01-01')] 

train.shape, val.shape, test.shape
```

- [ ] Begin with baselines for classification.

```python
# Baselines for classification

target = 'Great'
y_train = train[target]
y_train.value_counts(normalize=True)
```

```python
y_train.mode()
```

```python
majority_class = y_train.mode()[0]
y_pred = [majority_class] * len(y_train)
y_pred[:5]
```

```python
# Training accuracy of predicting majority class baseline
from sklearn.metrics import accuracy_score

accuracy_score(y_train, y_pred)
```

```python
# Calculate validation

y_val = val[target]
y_pred = [majority_class] * len(y_val)

accuracy_score(y_val, y_pred)
```

```python
# Looking at features to include

train.describe()
```

```python
# 1. Import estimator class
from sklearn.linear_model import LinearRegression

# 2. Instantiate this class
linear_reg = LinearRegression()

# 3. Arrange X feature matrices 
features = ['Cost', 'Length','Circum', 'Uniformity']
X_train = train[features]
X_val = val[features]

# Impute missing values
from sklearn.impute import SimpleImputer
imputer = SimpleImputer()
X_train_imputed = imputer.fit_transform(X_train)
X_val_imputed = imputer.transform(X_val)

# 4. Fit the model
linear_reg.fit(X_train_imputed, y_train)

# 5. Apply the model to new data.
# The predictions look like this ...
linear_reg.predict(X_val_imputed)
```

```python
# Coefficients

pd.Series(linear_reg.coef_, features)
```

- [ ] Use scikit-learn for logistic regression.
- [ ] Get your model's validation accuracy. (Multiple times if you try multiple iterations.)

```python
# Validation accuracy
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression(solver='lbfgs')
log_reg.fit(X_train_imputed, y_train)
print('Validation Accuracy', log_reg.score(X_val_imputed, y_val))
```
