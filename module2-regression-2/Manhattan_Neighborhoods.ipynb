```
import pandas as pd
import googlemaps
import numpy as np
```

```
# Fill in google maps API key to us the API. That will require open an account
# with Google, create a project and enable geolocation and geocode APIs, and
# obtain an API key.
API_key = 'yourAPIkeyhere'
gmaps = googlemaps.Client(key=API_key)
```

```
# List of neighborhoods
hoods = ["Chelsea", "Washington Heights", "Harlem", 
                   "East Harlem", "Upper West Side", 
                   "Upper East Side", "Midtown West", "Midtown East",
                   "Greenwich Village",
                   "Lower East Side", "Murray Hill",
                   "Stuyvesant Town", "Upper Manhattan", "Hell's Kitchen", 
                   "East Village", "SoHo", "Financial District", 
                   "Garment District", "Morningside Heights", "Tribeca",
                   "Chinatown", "Times Square"]
```

```
# Construct a dataframe for them. Latitudes and longitudes will be NaNs as 
# placeholders, to be filled in later.
neighborhoods = pd.DataFrame(hoods, columns=['neighborhood'])
neighborhoods['latitude'] = np.NaN
neighborhoods['longitude'] = np.NaN
neighborhoods
```

```
# For each neighborhood, find the aproximate coordinates for each and replace
# NaNs with values
for row in neighborhoods.index:
  
  # location searched for is '<neighborhood>, Manhattan'
  nhood = neighborhoods.loc[row]['neighborhood'] + ', Manhattan'

  # find the geocode data for it
  geocode_result = gmaps.geocode(nhood)

  # strip out the appropriate values from the API output
  geocode = geocode_result[0]['geometry']['location']

  # Update row in the dataframe
  neighborhoods.at[row, 'latitude'] = geocode['lat']
  neighborhoods.at[row, 'longitude'] = geocode['lng']

neighborhoods
```

```
# Output our CSV file!
neighborhoods.to_csv('manhattan_neighborhoods.csv')
```
